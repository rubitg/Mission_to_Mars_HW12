```python
# get in to the main web site...
url_main="https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars"
response = requests.get(url_main)
response
soup = BeautifulSoup(response.text, 'lxml')
soup
```

```python
# Main link to web page
main_link='https://astrogeology.usgs.gov'
main_link
```

```python
#Image 1
# https://astrogeology.usgs.gov/search/map/Mars/Viking/cerberus_enhanced   
#.. manually inspect the web page...
soup.find_all('div', class_="item")[0]
pre_img1 = soup.find('div', class_="item").a['href']
pre_img1
hemis_1=main_link+pre_img1
print(hemis_1)
```

```python

#Extract second item
soup.find_all('div', class_="item")[1]
pre_img2 = soup.find_all('div', class_="item")[1]
pre1_img2=pre_img2.a['href']
pre1_img2
# get link image 2
hemis_2=main_link+pre1_img2
print(hemis_2)
```

```python
#"image 3"
#https://astrogeology.usgs.gov/search/map/Mars/Viking/syrtis_major_enhanced
soup.find_all('div', class_="item")[2]
pre_img3 = soup.find_all('div', class_="item")[2]
pre1_img3=pre_img3.a['href']
pre1_img3
# get link image 2
hemis_3=main_link+pre1_img3
print(hemis_3)

```
